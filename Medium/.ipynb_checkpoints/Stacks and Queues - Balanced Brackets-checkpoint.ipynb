```python
def isBalanced(s):
    stack = []
    
    for i in s:
        if i in ('{', '[', '('):
            stack.append(i)
        elif len(stack) == 0:
            return 'NO'
        else:
            if i == '}':
                if stack[-1] != '{':
                    return 'NO'
                else:
                    stack.pop()
            elif i == ')':
                if stack[-1] != '(':
                    return 'NO'
                else:
                    stack.pop()
            elif i == ']':
                if stack[-1] != '[':
                    return 'NO'
                else:
                    stack.pop()
    if len(stack) == 0:
        return 'YES'
    else:
        return 'NO'
```
